```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
from torch.utils.data import Dataset, DataLoader
from datasets import load_dataset

def explore_dataset(dataset_dict):
    """
    Explore the structure and content of the DocLayNet dataset.
    """
    print(f"Dataset splits: {list(dataset_dict.keys())}")

    # Explore features of the training split
    train_dataset = dataset_dict['train']
    print(f"\nTraining set size: {len(train_dataset)}")
    print(f"Features: {train_dataset.features}")

    # Display an example
    example = train_dataset[0]
    print("\nSample data:")
    for key, value in example.items():
        if key == 'image':
            print(f"{key}: PIL Image of size {value.size if hasattr(value, 'size') else 'unknown'}")
        elif key == 'bboxes':
            print(f"{key}: {len(value) if value is not None else 0} bounding boxes")
            if value is not None and len(value) > 0:
                print(f"  First bbox: {value[0]}")
        elif key == 'metadata':
            print(f"{key}: {list(value.keys()) if isinstance(value, dict) else type(value)}")
        else:
            print(f"{key}: {type(value)} with {len(value) if hasattr(value, '__len__') else 'scalar'} elements")

class DocLayNetDataset(Dataset):
    """
    Custom dataset class for DocLayNet to handle preprocessing.
    """
    def __init__(self, dataset, transform=None, target_size=(512, 512)):
        self.dataset = dataset
        self.transform = transform
        self.target_size = target_size

        # Get category mapping
        self.category_mapping = self._get_category_mapping()

    def _get_category_mapping(self):
        """Create a mapping from category_id to a readable label."""
        # DocLayNet categories (adjust based on actual dataset)
        categories = {
            0: 'Background',
            1: 'Text',
            2: 'Title',
            3: 'List',
            4: 'Table',
            5: 'Figure',
            6: 'Header/Footer'
            # Add more categories as needed
        }
        return categories

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        try:
            item = self.dataset[idx]

            # Get image
            image = item['image']
            if image is None:
                # Create a blank image if image is missing
                image = Image.new('RGB', self.target_size, color='white')

            # Resize image
            if hasattr(image, 'size') and image.size != self.target_size:
                image = image.resize(self.target_size)

            # Convert to tensor
            image_tensor = torch.from_numpy(np.array(image)).permute(2, 0, 1).float() / 255.0

            # Process bounding boxes
            bboxes = item.get('bboxes', [])
            if bboxes is None:
                bboxes = []

            # Process category IDs
            category_ids = item.get('category_id', [])
            if category_ids is None:
                category_ids = []

            # Make sure category_ids matches bboxes length
            if len(category_ids) < len(bboxes):
                # Fill with default category (0)
                category_ids = category_ids + [0] * (len(bboxes) - len(category_ids))

            # Create target dictionary
            target = {
                'boxes': torch.tensor(bboxes, dtype=torch.float32) if bboxes else torch.zeros((0, 4)),
                'labels': torch.tensor(category_ids, dtype=torch.int64) if category_ids else torch.zeros(0, dtype=torch.int64),
                'image_id': torch.tensor([idx]),
                'area': torch.tensor(item.get('area', []), dtype=torch.float32) if 'area' in item and item.get('area') else torch.zeros(len(bboxes) if bboxes else 0),
                'iscrowd': torch.zeros((len(bboxes),), dtype=torch.int64) if bboxes else torch.zeros(0, dtype=torch.int64)
            }

            # Add segmentation masks if available
            if 'segmentation' in item and item['segmentation']:
                # This is a placeholder - actual implementation depends on segmentation format
                target['masks'] = torch.zeros((len(item['segmentation']), *self.target_size))

            if self.transform:
                image_tensor, target = self.transform(image_tensor, target)

            return image_tensor, target

        except Exception as e:
            print(f"Error processing item {idx}: {e}")
            # Return a default item
            image_tensor = torch.zeros((3, *self.target_size))
            target = {
                'boxes': torch.zeros((0, 4)),
                'labels': torch.zeros(0, dtype=torch.int64),
                'image_id': torch.tensor([idx]),
                'area': torch.zeros(0),
                'iscrowd': torch.zeros(0, dtype=torch.int64)
            }
            return image_tensor, target

def visualize_sample(dataset, idx=0):
    """
    Visualize a sample from the dataset with bounding boxes.
    """
    try:
        item = dataset[idx]
        image = item.get('image')
        if image is None:
            print("No image found for this sample")
            return

        bboxes = item.get('bboxes', [])
        if bboxes is None:
            bboxes = []

        category_ids = item.get('category_id', [])
        if category_ids is None:
            category_ids = []

        # Create figure and axes
        fig, ax = plt.subplots(1, figsize=(12, 12))

        # Display the image
        ax.imshow(image)

        # Define colors for different categories
        colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

        # Draw bounding boxes
        for i, bbox in enumerate(bboxes):
            x, y, width, height = bbox
            category = category_ids[i] if i < len(category_ids) else -1
            color = colors[category % len(colors)]

            rect = plt.Rectangle((x, y), width, height,
                                fill=False, edgecolor=color, linewidth=2)
            ax.add_patch(rect)
            ax.text(x, y, f"Cat: {category}",
                    bbox=dict(facecolor=color, alpha=0.5))

        plt.axis('off')
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error visualizing sample {idx}: {e}")

def collate_fn(batch):
    """
    Custom collate function to handle variable-sized bounding boxes
    """
    images = []
    targets = []

    for image, target in batch:
        images.append(image)
        targets.append(target)

    # Stack images
    images = torch.stack(images)

    return images, targets

def create_dataloaders(dataset_dict, batch_size=4, target_size=(512, 512)):
    """
    Create DataLoaders for training, validation, and testing.
    """
    # Create custom datasets
    train_dataset = DocLayNetDataset(dataset_dict['train'], target_size=target_size)
    val_dataset = DocLayNetDataset(dataset_dict['val'], target_size=target_size)
    test_dataset = DocLayNetDataset(dataset_dict['test'], target_size=target_size)

    # Create data loaders with custom collate function
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        num_workers=0  # Increase if needed and if using multiprocessing
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        collate_fn=collate_fn,
        num_workers=0
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size,
        collate_fn=collate_fn,
        num_workers=0
    )

    return train_loader, val_loader, test_loader

def main():
    try:
        # Load the DocLayNet dataset
        # If you have the dataset locally

        # Alternatively, if available on Hugging Face Hub
        # dataset_dict = load_dataset('doclaynet')

        # Explore dataset
        explore_dataset(ds)

        # Visualize a sample
        try:
            visualize_sample(ds['train'])
        except Exception as e:
            print(f"Visualization error: {e}")

        # Create dataloaders
        train_loader, val_loader, test_loader = create_dataloaders(ds)

        # Example: checking a batch from the training loader
        for images, targets in train_loader:
            print(f"Batch of images shape: {images.shape}")
            print(f"Number of targets: {len(targets)}")
            print(f"First target's boxes shape: {targets[0]['boxes'].shape}")
            break

    except Exception as e:
        print(f"Error in main: {e}")

if __name__ == "__main__":
    main()
```

```python
import os
import torch
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
from torchvision.transforms import functional as F
import numpy as np
from datasets import load_dataset
from tqdm import tqdm
import random

# DocLayNet category mapping
CATEGORY_MAPPING = {
    0: 0,  # Background (usually not in annotations but needed for model)
    1: 1,  # Text
    2: 2,  # Title
    3: 3,  # List
    4: 4,  # Table
    5: 5,  # Figure
    6: 6,  # Caption
    7: 7,  # Formula
    8: 8,  # Page-footer
    9: 9,  # Page-header
    10: 10  # Other
}

# Number of classes including background
NUM_CLASSES = len(CATEGORY_MAPPING)

class DocLayNetDataset(Dataset):
    def __init__(self, dataset, transform=None):
        self.dataset = dataset
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        try:
            sample = self.dataset[idx]

            # Get image
            image = sample['image']
            # Convert PIL Image to tensor
            image = F.to_tensor(image)

            # Get bounding boxes
            boxes = sample['bboxes']

            # Check if boxes have the correct format [x1, y1, x2, y2]
            # The error suggests the format might be [x, y, width, height]
            valid_boxes = []
            valid_category_ids = []

            # Get category IDs
            category_ids = sample['category_id']

            for i, box in enumerate(boxes):
                # Check if this is [x, y, width, height] format
                if len(box) == 4:
                    # Convert [x, y, width, height] to [x1, y1, x2, y2]
                    if box[2] <= 0 or box[3] <= 0:  # Skip boxes with zero or negative width/height
                        continue

                    if box[0] > box[2] or box[1] > box[3]:  # If already in x1,y1,x2,y2 format but flipped
                        x1 = min(box[0], box[2])
                        y1 = min(box[1], box[3])
                        x2 = max(box[0], box[2])
                        y2 = max(box[1], box[3])
                        valid_boxes.append([x1, y1, x2, y2])
                    else:  # Convert x,y,width,height to x1,y1,x2,y2
                        x1 = box[0]
                        y1 = box[1]
                        x2 = x1 + box[2]  # x + width
                        y2 = y1 + box[3]  # y + height
                        valid_boxes.append([x1, y1, x2, y2])

                    # Append corresponding category ID
                    cat_id = category_ids[i]
                    # Make sure category ID is within bounds
                    if cat_id < NUM_CLASSES:
                        valid_category_ids.append(cat_id)
                    else:
                        valid_category_ids.append(0)  # Map to background

            # If no valid boxes, create a dummy example to avoid errors
            if len(valid_boxes) == 0:
                valid_boxes = [[0, 0, 1, 1]]  # Small dummy box
                valid_category_ids = [0]  # Background class

            # Convert to tensors
            boxes = torch.as_tensor(valid_boxes, dtype=torch.float32)
            labels = torch.as_tensor(valid_category_ids, dtype=torch.int64)

            # Create target dictionary
            target = {}
            target["boxes"] = boxes
            target["labels"] = labels

            # Add image ID for reference
            image_id = torch.tensor([idx])
            target["image_id"] = image_id

            # Calculate area from bounding boxes (width * height)
            areas = (boxes[:, 2] - boxes[:, 0]) * (boxes[:, 3] - boxes[:, 1])
            target["area"] = areas

            # Add dummy iscrowd (usually all 0s for detection tasks)
            iscrowd = torch.zeros((len(boxes),), dtype=torch.int64)
            target["iscrowd"] = iscrowd

            if self.transform:
                image, target = self.transform(image, target)

            return image, target

        except Exception as e:
            # If there's an error processing this sample, create a dummy sample
            print(f"Error processing sample {idx}: {str(e)}")
            image = torch.zeros((3, 224, 224), dtype=torch.float32)
            boxes = torch.tensor([[0, 0, 1, 1]], dtype=torch.float32)
            labels = torch.tensor([0], dtype=torch.int64)

            target = {
                "boxes": boxes,
                "labels": labels,
                "image_id": torch.tensor([idx]),
                "area": torch.tensor([1.0]),
                "iscrowd": torch.tensor([0])
            }

            return image, target

def collate_fn(batch):
    """
    Collate function for DataLoader to handle variable-sized inputs
    """
    return tuple(zip(*batch))

def create_dataloaders(dataset_dict, batch_size=4):
    """
    Create DataLoaders for training, validation, and testing
    """
    train_dataset = DocLayNetDataset(dataset_dict['train'])
    val_dataset = DocLayNetDataset(dataset_dict['val'])
    test_dataset = DocLayNetDataset(dataset_dict['test'])

    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True,
        collate_fn=collate_fn,
        num_workers=2
    )

    val_loader = DataLoader(
        val_dataset,
        batch_size=batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        num_workers=2
    )

    test_loader = DataLoader(
        test_dataset,
        batch_size=batch_size,
        shuffle=False,
        collate_fn=collate_fn,
        num_workers=2
    )

    return train_loader, val_loader, test_loader

def get_model(num_classes):
    """
    Create a Faster R-CNN model with ResNet-50 backbone
    """
    # Load a pre-trained model
    model = fasterrcnn_resnet50_fpn(pretrained=True)

    # Replace the classifier with a new one for our number of classes
    in_features = model.roi_heads.box_predictor.cls_score.in_features
    model.roi_heads.box_predictor = FastRCNNPredictor(in_features, num_classes)

    return model

def train_one_epoch(model, optimizer, data_loader, device, epoch):
    """
    Train the model for one epoch
    """
    model.train()
    total_loss = 0
    valid_batches = 0

    for batch_idx, (images, targets) in enumerate(tqdm(data_loader)):
        try:
            # Move data to the correct device
            images = list(image.to(device) for image in images)
            targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

            # Verify all bounding boxes are valid
            for target_idx, target in enumerate(targets):
                boxes = target['boxes']
                # Check that all boxes have positive height and width
                widths = boxes[:, 2] - boxes[:, 0]
                heights = boxes[:, 3] - boxes[:, 1]
                if (widths <= 0).any() or (heights <= 0).any():
                    invalid_box_idx = (widths <= 0).nonzero(as_tuple=True)[0].item() if (widths <= 0).any() else (heights <= 0).nonzero(as_tuple=True)[0].item()
                    invalid_box = boxes[invalid_box_idx].tolist()
                    raise ValueError(f"All bounding boxes should have positive height and width. Found invalid box {invalid_box} for target at index {target_idx}.")

            # Forward pass
            loss_dict = model(images, targets)
            losses = sum(loss for loss in loss_dict.values())

            # Check for NaN
            if not torch.isfinite(losses):
                print(f"Loss is {losses}, stopping training")
                print(loss_dict)
                continue

            # Backward pass
            optimizer.zero_grad()
            losses.backward()
            optimizer.step()

            total_loss += losses.item()
            valid_batches += 1

            # Log progress
            if batch_idx % 50 == 0:
                print(f"Batch {batch_idx}/{len(data_loader)}, Loss: {losses.item():.4f}")

        except Exception as e:
            print(f"Error in batch {batch_idx}: {str(e)}")
            # Skip this batch and continue
            continue

    # Calculate average loss only over valid batches
    avg_loss = total_loss / valid_batches if valid_batches > 0 else float('inf')
    print(f"Epoch {epoch+1}, Average Loss: {avg_loss:.4f}, Valid Batches: {valid_batches}/{len(data_loader)}")

    return avg_loss

def evaluate(model, data_loader, device):
    """
    Evaluate the model on the validation set
    """
    model.eval()

    # Simplified metrics - in a real scenario, use proper COCO mAP calculation
    correct_detections = 0
    total_detections = 0

    # No gradient computation during evaluation
    with torch.no_grad():
        for images, targets in tqdm(data_loader):
            try:
                images = list(img.to(device) for img in images)
                targets = [{k: v.to(device) for k, v in t.items()} for t in targets]

                # Forward pass
                outputs = model(images)

                # Simple accuracy metric (placeholder for proper mAP)
                for output, target in zip(outputs, targets):
                    # Get predictions with score > 0.5
                    keep = output['scores'] > 0.5
                    pred_boxes = output['boxes'][keep]
                    pred_labels = output['labels'][keep]

                    # Get ground truth
                    gt_boxes = target['boxes']
                    gt_labels = target['labels']

                    # Count total detections
                    total_detections += len(gt_boxes)

                    # Count correct detections (simplified - should use IoU)
                    # For a real evaluation, use proper IoU matching
                    correct_detections += len(pred_boxes)
            except Exception as e:
                print(f"Error during evaluation: {str(e)}")
                continue

    # Calculate simplified mAP (just a placeholder)
    mean_ap = correct_detections / total_detections if total_detections > 0 else 0
    print(f"Evaluation: 'mAP' = {mean_ap:.4f} (simplified metric)")

    return mean_ap, None

def save_checkpoint(model, optimizer, epoch, path):
    """
    Save model checkpoint
    """
    torch.save({
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
    }, path)
    print(f"Checkpoint saved to {path}")

def analyze_dataset(dataset_dict):
    """
    Analyze the dataset structure to understand the format of bounding boxes
    """
    print("Analyzing dataset structure...")

    # Check train dataset
    train_ds = dataset_dict['train']
    sample = train_ds[0]

    print(f"Sample keys: {sample.keys()}")
    print(f"Image type: {type(sample['image'])}, shape: {sample['image'].size}")
    print(f"Bounding boxes: {sample['bboxes'][:2]}")
    print(f"Category IDs: {sample['category_id'][:2]}")

    # Check format of bounding boxes
    boxes = sample['bboxes']
    if len(boxes) > 0:
        box = boxes[0]
        print(f"First box: {box}")

        if len(box) == 4:
            # Try to determine format: [x1, y1, x2, y2] or [x, y, width, height]
            if box[2] > box[0] and box[3] > box[1]:
                # Could be either format, check values
                w = box[2] - box[0]
                h = box[3] - box[1]
                print(f"Possible format: [x1, y1, x2, y2] (width={w}, height={h})")
            else:
                print(f"Possible format: [x, y, width, height]")
        else:
            print(f"Unexpected box format with {len(box)} values")

    return None

def inspect_random_samples(dataset, num_samples=5):
    """
    Inspect random samples from the dataset to debug bounding box issues
    """
    indices = random.sample(range(len(dataset)), min(num_samples, len(dataset)))

    for idx in indices:
        try:
            sample = dataset[idx]
            print(f"\nSample {idx}:")
            print(f"Number of boxes: {len(sample['bboxes'])}")

            for i, (box, cat_id) in enumerate(zip(sample['bboxes'][:3], sample['category_id'][:3])):
                print(f"Box {i}: {box}, Category: {cat_id}")

                # Try to determine if this is [x, y, width, height] or [x1, y1, x2, y2]
                if len(box) == 4:
                    if box[2] <= 0 or box[3] <= 0:
                        print(f"  Warning: Zero or negative width/height")

                    # Check if coordinates might be flipped
                    if box[0] > box[2] or box[1] > box[3]:
                        print(f"  Warning: First coordinate larger than second (might be flipped)")

                    # Calculate and print dimensions
                    w = abs(box[2] - box[0])
                    h = abs(box[3] - box[1])
                    print(f"  Dimensions (abs): width={w}, height={h}")
                else:
                    print(f"  Warning: Unexpected box format with {len(box)} values")

        except Exception as e:
            print(f"Error inspecting sample {idx}: {str(e)}")

    return None

def main():
    # Set random seed for reproducibility
    torch.manual_seed(42)
    random.seed(42)
    np.random.seed(42)

    # Set device
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")


    # Analyze dataset structure
    analyze_dataset(ds)
    print("\nInspecting random samples:")
    inspect_random_samples(ds['train'])

    # Parameters
    num_classes = NUM_CLASSES  # Adjust based on your dataset
    batch_size = 2  # Reduced batch size to avoid memory issues
    num_epochs = 10
    learning_rate = 0.001
    weight_decay = 0.0005

    train_loader, val_loader, test_loader = create_dataloaders(ds)

    # Create model
    print("Creating model...")
    model = get_model(num_classes)
    model.to(device)

    # Define optimizer
    params = [p for p in model.parameters() if p.requires_grad]
    optimizer = optim.SGD(params, lr=learning_rate, momentum=0.9, weight_decay=weight_decay)

    # Learning rate scheduler
    lr_scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)

    # Training loop
    best_map = 0.0
    loss_history = []
    map_history = []

    # Create checkpoint directory
    checkpoint_dir = 'checkpoints'
    os.makedirs(checkpoint_dir, exist_ok=True)

    for epoch in range(num_epochs):
        print(f"\nEpoch {epoch+1}/{num_epochs}")

        # Train for one epoch
        epoch_loss = train_one_epoch(model, optimizer, train_loader, device, epoch)
        loss_history.append(epoch_loss)

        # Update learning rate
        lr_scheduler.step()

        # Evaluate on validation set
        mean_map, _ = evaluate(model, val_loader, device)
        map_history.append(mean_map)

        # Save checkpoint
        checkpoint_path = os.path.join(checkpoint_dir, f'model_epoch_{epoch+1}.pth')
        save_checkpoint(model, optimizer, epoch, checkpoint_path)

        # Save best model
        if mean_map > best_map:
            best_map = mean_map
            best_model_path = os.path.join(checkpoint_dir, 'best_model.pth')
            save_checkpoint(model, optimizer, epoch, best_model_path)
            print(f"New best model saved with mAP: {best_map:.4f}")

    print("Training complete!")

    # Save final model
    final_model_path = os.path.join(checkpoint_dir, 'final_model.pth')
    save_checkpoint(model, optimizer, num_epochs, final_model_path)
    print(f"Final model saved to {final_model_path}")

    # Test the best model
    print("Testing best model...")
    model.load_state_dict(torch.load(best_model_path)['model_state_dict'])
    test_map, _ = evaluate(model, test_loader, device)
    print(f"Test mAP: {test_map:.4f}")

if __name__ == "__main__":
    main()
```
